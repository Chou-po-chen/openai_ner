```python
import os 
import numpy as np
import pandas as pd
```

```python
# DATA_DIR   = rf'..\First_Phase\data'
# ANSWER_DIR = rf'..\First_Phase\answer'
# ANS_TXT_FILE = rf'..\First_Phase\answer.txt'

DATA_DIR   = rf'..\merge_first_second_phase\data'
ANSWER_DIR = rf'..\merge_first_second_phase\answer'
ANS_TXT_FILE = rf'..\merge_first_second_phase\answer.txt'


names=['file','label','start','end','content','time']
df = pd.read_csv(ANS_TXT_FILE , sep='\t', header=None , names=names)

# marge time
df.loc[df['time'].notna() , 'content'] = df['content'] + '=>' + df['time'] 


df.head(10)
```

```python
file_listdir = os.listdir(DATA_DIR)
all_file_names = []
for file_name in file_listdir:
    if file_name.endswith(".txt"):  # Filter files with a .txt extension
        all_file_names.append(file_name)
# print(all_file_names)

# ### seperate data
# ch = all_file_names[:300] + all_file_names[-300:]
# all_file_names = ch
# all_file_names

# ### sel data from must_choose_data
choose_file = []
with open('../lab/must_choose_file_list.txt' , 'r') as f:
    for line in f.readlines():
        line = line.strip()
        if len(line) > 0:
            choose_file.append(line+'.txt')
all_file_names = choose_file
all_file_names
```

```python
for file in all_file_names:
    file_path = os.path.join(DATA_DIR, file)
    file_name = file.replace('.txt','')
    print(file_path)
    new_df = df.loc[df['file'] == file_name , ['label','content']]
    print(new_df)
    # new_df.to_csv(os.path.join(ANSWER_DIR, f'{file_name}_PHI.txt'), sep='\t', index=False, header=False)
    # write to file ex format is MEDICALRECORD: 091016.NMT
    with open(os.path.join(ANSWER_DIR, f'{file_name}_PHI.txt'), 'w') as f:
        for index, row in new_df.iterrows():
            f.write(f'{row["label"]}: {row["content"]}\n')
```

```python
def gen_data_dict(user,assistant):
    data = {}
    data['messages'] = []
    data['messages'].append({'role': 'system'   ,   'content': 'Find PHI(DOCTOR,DATE,IDNUM,MEDICALRECORD,PATIENT,HOSPITAL,TIME,DEPARTMENT,CITY,ZIP,STREET,STATE,AGE,ORGANIZATION,DURATION,PHONE,URL,LOCATION-OTHER,SET,COUNTRY,ROOM) and normalize time information'})
    data['messages'].append({'role': 'user'     ,   'content': user})
    data['messages'].append({'role': 'assistant',   'content': assistant})
    return data
```
