```python
### amswer.txt reader 
# DATA_DIR   = rf'..\First_Phase\data'
# ANSWER_DIR = rf'..\First_Phase\answer'
import pandas as pd
# ANS_TXT_FILE = rf'..\First_Phase\answer.txt'
# ANS_TXT_FILE = rf'..\First_Phase_Validation\answer.txt'
ANS_TXT_FILE = rf'..\merge_first_second_phase\answer.txt'

names=['file','label','start','end','content','time']
df = pd.read_csv(ANS_TXT_FILE , sep='\t', header=None , names=names)

# marge time
df.loc[df['time'].notna() , 'content'] = df['content'] + '=>' + df['time'] 
```

```python
res = df['label'].value_counts()
# must choose label is count<100
must_choose_label = res[res<150].index.tolist()
print(res)
print(must_choose_label)

# if file have must_choose_label , add to must_choose_file  , delete same item
must_choose_file = df[df['label'].isin(must_choose_label)]['file'].tolist()
print(must_choose_file)

must_choose_file = list(set(must_choose_file))
print(must_choose_file)
with open('must_choose_file_list.txt', 'w') as f:
    f.write('\n'.join(must_choose_file))
```
